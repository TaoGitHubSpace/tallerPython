```python
f2=["a","b","c","d","e","f","g","h","i"] # Lista original
Used=["n","n","n","n","n","n","n","n","n"] # Lista de transferencia de posiciones
Tie=["a","b","c","d","e","f","g","h","i"] # Lista de comparación de empate

nam1=input("Jugador 1 (X): Escribe tu nombre\n") # Definir nombre del jugador 1 y el caracter que va a usar
nam2=input("Jugador 2 (O): Escribe tu nombre\n") # Definir nombre del jugador 1 y el caracter que va a usar

print("[",f2[0],f2[1],f2[2],"]"), # Imprimir matriz visual inicial
print("[",f2[3],f2[4],f2[5],"]"),
print("[",f2[6],f2[7],f2[8],"]")

while True: # Bucle de ganador o empate
    # Condicional de empate
    if (f2[0] and f2[1] and f2[2] and f2[3] and f2[4] and f2[5] and f2[6] and f2[7] and f2[8])==("X" or "O"):
        print("!OH NO¡, Empataron")
        break # Fin del bucle para empate
    while True: # Bucle de asignación de posición del jugador 1
        print(nam1," por favor elija una ubicación (X)\n") # texto de solicitud de dato con nombre
        U1=input().lower() # Asignación de variable al dato solicitado al jugador 1, en minúscula
        if U1 in f2: # Condicional si la variable existe en la lista original
            Used[f2.index(U1)]=U1 # Guardar en lista de transferencia el dato en posición de la lista original
            f2[f2.index(U1)]="X" # Asignación de representación de jugador 1 en lista original
            break # Salida de bucle de asignación de posición del jugador 1
        else:
            print("Ubicación inexistente") # Impresión en caso de repetir ubicación o asignar una letra diferente
        
    print("[",f2[0],f2[1],f2[2],"]"), # Imprimir matriz visual después de cambio
    print("[",f2[3],f2[4],f2[5],"]"),
    print("[",f2[6],f2[7],f2[8],"]")
    
    V1=f2[0]==f2[1]==f2[2] # Posibles salidas de victoria
    V2=f2[3]==f2[4]==f2[5]
    V3=f2[6]==f2[7]==f2[8]
    V4=f2[0]==f2[3]==f2[6]
    V5=f2[1]==f2[4]==f2[7]
    V6=f2[2]==f2[5]==f2[8]
    V7=f2[0]==f2[4]==f2[8]
    V8=f2[2]==f2[4]==f2[6]
    if (V1 or V2 or V3 or V4 or V5 or V6 or V7 or V8)== True: # Condicional de victoria para el Jugador 1
        print("Felicitaciones", nam1, "haz ganado") # Impresión para el ganador
        print("Lo siento", nam2, "perdiste") # Impresión para el perdedor
        break # Salida de bucle de ganador del jugador 1
    elif Tie == Used: # Condicional para definir empate
        print("!OH NO¡, Empataron") # impresión de True para condicional de empate
        break # Salida de bucle de empate
    
    while True:
        print(nam2," por favor elija una ubicación (O)\n")
        U2=input().lower()
        if U2 in f2:
            Used[f2.index(U2)]=U2
            f2[f2.index(U2)]="O"
            break
        else:
            print("Ubicación inexistente")
        
    print("[",f2[0],f2[1],f2[2],"]"),
    print("[",f2[3],f2[4],f2[5],"]"),
    print("[",f2[6],f2[7],f2[8],"]")
    
    V1=f2[0]==f2[1]==f2[2]
    V2=f2[3]==f2[4]==f2[5]
    V3=f2[6]==f2[7]==f2[8]
    V4=f2[0]==f2[3]==f2[6]
    V5=f2[1]==f2[4]==f2[7]
    V6=f2[2]==f2[5]==f2[8]
    V7=f2[0]==f2[4]==f2[8]
    V8=f2[2]==f2[4]==f2[6]
    if (V1 or V2 or V3 or V4 or V5 or V6 or V7 or V8)== True:
        print("Felicitaciones", nam2, "haz ganado")
        print("Lo siento", nam1, "perdiste")        
        break
    elif Tie == Used:
        print("!OH NO¡, Empataron")
        break
```
