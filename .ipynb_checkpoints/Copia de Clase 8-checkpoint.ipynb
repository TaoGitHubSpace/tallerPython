```python
# Taller organizar
# Ordenar de menor a mayor (Ok)
# Ordenar de mayor a menor (Ok)
# Borrar posiciones impares 
# Borrar posiciones pares
# Borrar números inpares
# Borrar números pares
L1=[1,3,7,-1,-2,0,15,42,37,2,4,-10] # lista inicial
L1mM=[] # futura lista de menor a Mayor
L1Mm=[] # futura lista de Mayor a menor
L2NoPosPar=[] # futura lista para poder eliminar posiciones pares de L1Mm sin afectar toda la estructura
L2NoPosImp=[] # futura lista para poder eliminar posiciones impares de L1Mm sin afectar toda la estructura
L3NoNumPar=[] # futura lista para poder eliminar numeros pares de L1Mm sin afectar toda la estructura
L3NoNumImp=[] # futura lista para poder eliminar numeros impares de L1Mm sin afectar toda la estructura

for a in range(len(L1)): # Creamos los valores a reemplazar durante la organización
    L1mM.append(a)
print("La lista original",L1)

ValMax=0 # Valor inicial de comparación para determinar el mayor
for dat in L1: # Para los objetos en la lista inicial evaluar
    if dat>ValMax: # Si el objeto es mayor que el valor inicial de comparación, entonces
        ValMax=dat # Reemplazamos el balor inicial por el objeto de comparación.

ValMin=ValMax # Valor inicial de comparación para determinar el menor
for dat in L1: # Para los objetos en la lista inicial evaluar
    if dat<ValMin: # Si el objeto es menor que el valor inicial de comparación, entonces
        ValMin=dat # Reemplazamos el balor inicial por el objeto de comparación.

# Una vez obtenido el máximo y el mínimo podemos empezar a organizar desde y hasta donde comparar

while True:
    if len(L1)==0: # Si la L1 se queda sin datos habremos terminado la organización 
        print("Lista de mayor a menor 'L1Mm':",L1Mm) 
        break
    ValMax=ValMin # El valor inicial de comparación debe ser el menor de la lista inicial. 
    for val in L1: # Para los valores de la Lista inicial, evaluar 
        if val>ValMax: # Si el valor es mayor al valor de comparación, entonces
            ValMax=val # El valor maximo ahora es el valor comparado de la lista inicial
    L1mM[len(L1)-1]=ValMax # Agregar el valor maximo en la última posición de la lista menor a Mayor (L1mM)
    L1Mm.append(ValMax) # Agregar el valor maximo en la lista de Mayor a menor (L1Mm)
    L2NoPosPar.append(ValMax) # Agregar el valor maximo en la lista de eliminar posiciones pares (L2NoPosPar)
    L2NoPosImp.append(ValMax) # Agregar el valor maximo en la lista de eliminar posiciones impares (L2NoPosImp)
    L3NoNumPar.append(ValMax) # Agregar el valor maximo en la lista de eliminar numeros pares (L3NoNumPar)
    L3NoNumImp.append(ValMax) # Agregar el valor maximo en la lista de eliminar numeros impares (L3NoNumImp)
    L1.remove(ValMax) # Eliminar dato de la lista original para poder reevaluar el nuevo numero mayor
print("Lista de menor a mayor 'L1mM':",L1mM)

PosPar=[] # Lista vacia para poder almacenar las posiciones pares de la lista original
for a in L2NoPosPar: # Para los datos de la lista evaluar
    if L2NoPosPar.index(a)%2==0: # Si la ubicación del dato dividido entre 2 deja un residuo igual a 0 entonces 
        PosPar.append(a) # Guardar en la lista vacia de posiciones pares
for b in PosPar: # Para los datos de la lista nueva de posiciones pares evaluar
    if b in L2NoPosPar: # si el dato evaluado se encuentra en la lista a evaluar, entonces
        L2NoPosPar.remove(b) # Eliminar dato de la lista a evaluar
print("Lista sin posiciones pares de 'L1Mm':",L2NoPosPar)

PosImp=[] # Lista vacia para poder almacenar las posiciones pares de la lista original
for a in L2NoPosImp: # Para los datos de la lista evaluar
    if L2NoPosImp.index(a)%2!=0: # Si la ubicación del dato dividido entre 2 deja un residuo diferente a 0 entonces 
        PosImp.append(a) # Guardar en la lista vacia de posiciones impares
for b in PosImp: # Para los datos de la lista nueva de posiciones impares evaluar
    if b in L2NoPosImp: # si el dato evaluado se encuentra en la lista a evaluar, entonces
        L2NoPosImp.remove(b) # Eliminar dato de la lista a evaluar
print("Lista sin posiciones impares de 'L1Mm':",L2NoPosImp)

NumPar=[] # Lista vacia para poder almacenar los numeros pares de la lista original
for a in L3NoNumPar: # Para los datos de la lista evaluar
    if a%2==0: # Si el numero dividido entre 2 deja un residuo igual a 0 entonces 
        NumPar.append(a) # Guardar en la lista vacia de números pares
for b in NumPar: # Para los datos de la lista nueva de numeros pares evaluar
    if b in L3NoNumPar: # si el dato evaluado se encuentra en la lista a evaluar, entonces
        L3NoNumPar.remove(b) # Eliminar dato de la lista a evaluar
print("Lista sin números pares de 'L1Mm':",L3NoNumPar)

NumImp=[] # Lista vacia para poder almacenar los numeros impares de la lista original
for a in L3NoNumImp: # Para los datos de la lista evaluar
    if a%2!=0: # Si el numero dividido entre 2 deja un residuo diferente a 0 entonces 
        NumImp.append(a) # Guardar en la lista vacia de números impares
for b in NumImp: # Para los datos de la lista nueva de numeros impares evaluar
    if b in L3NoNumImp: # si el dato evaluado se encuentra en la lista a evaluar, entonces
        L3NoNumImp.remove(b) # Eliminar dato de la lista a evaluar
print("Lista sin números impares de 'L1Mm':",L3NoNumImp)
```

```python
B=["A","a","Z","z","h","c","g","i","G"]
mM=[]
Mm=[]
Lel=[]
print("La lista original",B)
for a in range(len(B)):
    mM.append(a)
max_="A" # Frente al ejercicio anterior este es el unico cambio para poder evaluar texto
for num in B:
    if num>max_:
        max_=num
min_=max_
for num in B:
    if num<min_:
        min_=num
while True:
    if len(B)==0:
        print("Lista de mayor a menor",Mm)
        break
    maxi=min_
    for num in B:
        if num>maxi:
            maxi=num
    mM[len(B)-1]=maxi
    Mm.append(maxi)
    Lel.append(maxi)
    B.remove(maxi)
print("Lista de menor a mayor",mM)

PosParL=[]
for a in mM:
    if mM.index(a)%2==0:
        PosParL.append(a)
for b in PosParL:
    if b in mM:
        mM.remove(b)
print(mM)

PosImpL=[]
for a in mM:
    if mM.index(a)%2!=0:
        PosImpL.append(a)
for b in PosImpL:
    if b in mM:
        mM.remove(b)
print(mM)

print("Cual letra quiere borrar")
Bor=input()
if Bor in Lel:
    Lel.remove(Bor)
print("Lista despues de eliminar:", Lel)
```
